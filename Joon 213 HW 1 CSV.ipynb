```python
import pandas as pd
```

```python
names = ['Serial No.','GRE Score', 'TOEFL Score', 'University Rating', 'SOP', 'LOR', 'CGPA', 'Research','Chance of Admit']
admission_df = pd.read_csv("C:\\Users\\jyhah\\Downloads\\admission_predict.csv", header = None, names = names)
```

```python
admission_df.head()
```

```python
#Dropping first row since it is a header
admission_df = admission_df.drop([0])
```

```python
#Setting columns as floats so the data could interpret correctly as integers. Used coerce function to drop any columns that
#could not be converted to numerical form.
cols = ['Serial No.','GRE Score', 'TOEFL Score', 'University Rating', 'SOP', 'LOR', 'CGPA', 'Research','Chance of Admit']
admission_df[cols] = admission_df[cols].apply(pd.to_numeric, errors='coerce')
```

```python
admission_df.head()
```

```python
admission_df.info()
```

```python
#Finding null columns and rows
admission_df.isnull()
```

```python
admission_df = admission_df.dropna()
admission_df.drop(['Serial No.'], axis = 1, inplace = True)
#Dropping Serial No. as it doesn't serve any purpose other than numbering
```

```python
admission_df.describe()
```

```python
%matplotlib inline
```

```python
q_df = admission_df.quantile([.25, .75])
q_df.loc['iqr'] = q_df.loc[0.75] - q_df.loc[0.25]
q_df.loc['whisker_length'] = 1.5 * q_df.loc['iqr']
q_df.loc['max_whisker'] = q_df.loc['whisker_length'] + q_df.loc[0.75]
q_df.loc['min_whisker'] = q_df.loc[0.25] - q_df.loc['whisker_length']
q_df
```

```python
import numpy as np
std_gre = admission_df['GRE Score'].std()
std_gre
```

```python
maxoutlierGRE = admission_df['GRE Score'].mean() + (std_gre * 3)
maxoutlierGRE
```

```python
#With this we can see that there are no outliers in GRE score.
#We can continue using max and min outlier tests on all of the variables and see if we could drop them.
std_uni = admission_df['University Rating'].std()
std_uni
maxoutlierUNI = admission_df['University Rating'].mean() + (std_uni * 3)
maxoutlierUNI
```
